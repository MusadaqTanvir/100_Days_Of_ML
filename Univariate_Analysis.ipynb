```python
import pandas as pd
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'Datasets/titanic_train.csv')
df.head()
```

## First Categorical Data

### a: Countplot

```python
sns.countplot(x=df['Survived'],data=df)
```

```python
# Numerical values 
df['Survived'].value_counts()
```

```python
# Checking the Pclasses on the ship
sns.countplot(x=df['Pclass'], data=df)
```

```python
# gender distribtution on the ship
sns.countplot(x=df['Sex'],data=df)
```

```python
# checking the embarked column from which port the most passenger boarded
sns.countplot(x=df['Embarked'],data=df)
```

## Pie chart

```python
# same questions but with the percentage requirements 
result = df['Survived'].value_counts()
plt.pie(result,labels={'Not Survived':0,'Survived':1},colors=sns.color_palette('dark'),autopct='%.0f%%')
plt.show()
```

```python
# classes percentage
classes = df['Pclass'].value_counts()
keys = {3:len(df[df['Pclass']==3]),1:len(df[df['Pclass']==2]),2:len(df[df['Pclass']==1])}
plt.pie(classes,labels=keys,colors=sns.color_palette('dark'),autopct='%.0f%%')
plt.show()
```

```python
gender = df['Sex'].value_counts()
plt.pie(gender,labels=['Male','Female'],colors=sns.color_palette('dark'),autopct='%.0f%%')
plt.show()
```

```python
# Embarked
Embarked = df['Embarked'].value_counts()
plt.pie(Embarked,labels=['S', 'C', 'Q'],colors=sns.color_palette('dark'),autopct='%.0f%%')
plt.show()
```

```python
# same thing can be done using pandas' plotting
df['Survived'].value_counts().plot(kind='pie',autopct='%.2f%%')
```
